```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2) #Merges intersection of two dataframes here key1 three values are common in both dataframes so ot merges both
#By default how="inner"
```

```python
pd.merge(data1,data2,how="left")#lEFT MEANS DATA1 THAT IS IT WILL KEEP ALL DATA OF DATA1 AMD THEN WHEREVER MATCH IT WILL PLACE DATA
```

```python
pd.merge(data1,data2,how="right")#RIGHT MEANS DATA2 THAT IS IT WILL KEEP ALL DATA OF DATA2 AMD THEN WHEREVER MATCH IT WILL PLACE DATA
```

```python
pd.merge(data1,data2,how = 'outer')#FOR ALL DATA1 AND DATA2 IT WILL KEEP
```

```python
pd.concat([data1,data2],axis=0)
```

```python
pd.merge(data1,data2,how = 'cross')#All possible combination of data1 and data2 therefore len(data1)=5 and len(data2)=5 =25 
#Therefore we have 25 elemnets in result
```

```python
pd.merge(data1,data2,how = 'left' , left_on='key2' , right_on= 'key4' )
```

```python
def func(df):
    l=df["Fare"].to_list()
    l1=[]
    for i in l:
        if (int(i)<10):
            l1.append("Low")
        elif (int(i)<20):
            l1.append("Medium")
        else:
            l1.append("High")
    df["Standard"]=l1
    return df
```
